```python
##data cleaning kaggle data

df2.duplicated().sum() #checking for duplicates #no duplicates

df2=df2.drop(columns=['Legislative District'])

minor_missing=['County', 'City', 'Postal Code', 'Electric Utility', '2020 Census Tract','Model'] #replacing the missing values in categorical columns with 'unknown'
df2[minor_missing]=df2[minor_missing].fillna('Unknown')

df2[' Electric Range']=df2['Electric Range'].fillna(df2['Electric Range'].median()) #filling numerical missing values with median
df2['Base MSRP']=df2['Base MSRP'].fillna(df2['Base MSRP'].median())


df2=df2.dropna(subset=['Vehicle Location']) ##removing rows with missing vehicle location

```
